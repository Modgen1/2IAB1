```python
import numpy as np
import pandas as pd
from PIL import Image
```

```python
with Image.open('linal.jpg') as img:
    img.show()
```

```python
img = Image.open('linal.jpg').convert('L')
arr = np.array(img)
u, s_vect, v_t = np.linalg.svd(arr, full_matrices=False)
v = v_t.T
s = np.diag(s_vect)
```

```python
size = img.size[0] * img.size[1]
print(f'{size / 1000000} MB')
```

```python
size_svd = u.shape[0] * u.shape[1] + s.shape[0] + v.shape[0] * v.shape[1]
print(f'{size_svd / 1000000} MB')
```

```python
pd.DataFrame(u)
```

```python
pd.DataFrame(s)
```

```python
pd.DataFrame(v)
```

```python
restored_arr = u @ np.diag(s) @ v_t
restored_img = Image.fromarray(restored_arr)
restored_img.show()
```

```python
u_100 = u[:, :100]
s_100 = s[:100, :100]
v_100 = v[:, :100]
```

```python
arr_100 = u_100 @ s_100 @ v_100.T
img_100 = Image.fromarray(arr_100).convert('RGB')
```

```python
img_100.show()
```

```python
size_100 = u_100.shape[0] * u_100.shape[1] + s_100.shape[0] + v_100.shape[0] * v_100.shape[1]
print(f'{size_100 / 1000000} MB')
```
